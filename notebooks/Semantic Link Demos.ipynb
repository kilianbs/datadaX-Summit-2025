#### Cambiar tema a todos los informes


```python
#theme_file_path = "/lakehouse/default/Files/Themes/BlueTheme.json"
theme_file_path = "/lakehouse/default/Files/Themes/BlackTheme.json"

# Obtener todos los reports del workspace
reports = fabric.list_reports(workspace=workspace)

# Itera por los reportes y aplica el tema
for report in reports["Name"].tolist():
    try:
        rpt = ReportWrapper(report=report, workspace=workspace)
        rpt.set_theme(theme_file_path=theme_file_path)
        print(f"✅ Tema aplicado al informe: {report}")
    except Exception as e:
        print(f"❌ Error al aplicar el tema al informe: {report} - Error: {str(e)}")
```

#### Modificar modelo funciones TOM

```python
from sempy_labs.tom import connect_semantic_model as csm
```

```python
#lista tablas
with csm(dataset=dataset, readonly=True, workspace=workspace) as tom:
    for t in tom.model.Tables:
            print(t.Name)
```

```python
#lista columnas
with csm(dataset=dataset, readonly=True, workspace=workspace) as tom:
    for t in tom.all_columns():
        print(t.Name)
```

```python
# set IsAvailableInMDX
# ejemplo 1: Para una columna
with csm(dataset=dataset, readonly=False, workspace=workspace) as tom:       
    tom.model.Tables['Country'].Columns['Id'].IsAvailableInMDX = False
```

```python
# set IsAvailableInMDX
# ejemplo 2: para columnas relevantes 
with csm(dataset=dataset, readonly=False, workspace=workspace) as tom:
    
    if not tom.is_direct_lake():
        for c in tom.all_columns():
            if c.IsAvailableInMDX and c.SortByColumn is None and not any(tom.used_in_sort_by(column=c)) and not any(tom.used_in_hierarchies(column=c)):
                c.IsAvailableInMDX = False
```

```python
# lista medidas
with csm(dataset=dataset, readonly=True, workspace=workspace) as tom:
    for t in tom.all_measures():
        print(t.Name)
```

## Modificación JSON de informes
#### Varios informes
Actualmente dicha solución no funciona con el nuevo modelo PBIR porque la API para obtener el JSON solo funciona con el PBIR legacy <br>
Mas info sobre el error: https://www.techtacofriday.com/the-tmsl-to-tmdl-switch-part2-i-did-my-homework/

```python
#nos guardamos los json de todos los informes
current_jsons = {}
for report_name in report_names:
    current_jsons[report_name] = rep.get_report_json(report_name)
```

```python
import json

# Reemplazamos nombre en el json
new_jsons = {}
for report_name, current_json in current_jsons.items():
    new_jsons[report_name] = json.loads(
        json.dumps(current_json)
            .replace("MProfit" , "MProfit2")
    )
print('Cambio realizado')
```

```python
for report_name, new_json in new_jsons.items():
    rep.update_report_from_reportjson(
        report = report_name,
        report_json = new_json
    )
    print(f"Informe actualizado: {report_name}")
```
